```python
# @title 1. Data Loading & Setup

import pandas as pd

df = pd.read_excel("/content/ELGI_SALES_DATA_POWERBI.xlsx")

print("Column Names:")
print(df.columns.tolist())
print()
print("First 10 Rows:")
df.head(10)
```

```python
# @title 2. Correlation Analysis (Pearson's r)
# H0: There is no significant relationship between Quantity Sold and Sales Revenue.
# H1: There is a significant relationship between Quantity Sold and Sales Revenue.

from scipy.stats import pearsonr

r, p = pearsonr(
    df['Total_Quantity_Sold'],
    df['Total_Sales_Value_(INR)']
)

print("=" * 45)
print("       CORRELATION ANALYSIS RESULTS")
print("=" * 45)
print(f"  Correlation Coefficient (r) : {round(r, 4)}")
print(f"  P-value                     : {round(p, 6)}")
print("-" * 45)

if p < 0.05:
    print("  Result        : Reject H0")
    print("  Interpretation: There is a statistically")
    print("  significant relationship between Quantity")
    print("  Sold and Sales Revenue.")
else:
    print("  Result        : Accept H0")
    print("  Interpretation: No statistically significant")
    print("  relationship found.")

print("=" * 45)
```

```python
# @title 3. One-Way ANOVA (Sales by Product Category)
# H0: There is no significant difference in sales revenue across product categories.
# H1: There is a significant difference in sales revenue across product categories.

from scipy.stats import f_oneway

groups = [
    group['Total_Sales_Value_(INR)'].dropna()
    for name, group in df.groupby('Product_Category')
]

f_stat, p_value = f_oneway(*groups)

print("=" * 45)
print("          ONE-WAY ANOVA RESULTS")
print("=" * 45)
print(f"  F-statistic : {round(f_stat, 4)}")
print(f"  P-value     : {round(p_value, 6)}")
print("-" * 45)

if p_value < 0.05:
    print("  Result        : Reject H0")
    print("  Interpretation: Significant difference in")
    print("  sales revenue exists across product categories.")
else:
    print("  Result        : Accept H0")
    print("  Interpretation: No significant difference")
    print("  in sales revenue across product categories.")

print("=" * 45)
```

```python
# @title 4. Simple Linear Regression (Quantity Sold → Sales Revenue)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

X = df[['Total_Quantity_Sold']]
y = df['Total_Sales_Value_(INR)']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("=" * 45)
print("      LINEAR REGRESSION RESULTS")
print("=" * 45)
print(f"  Intercept   : {round(model.intercept_, 2)}")
print(f"  Coefficient : {round(model.coef_[0], 2)}")
print(f"  R-squared   : {round(r2_score(y, y_pred), 4)}")
print("-" * 45)
print("  Finding:")
print("  Quantity sold has a positive impact on")
print("  sales revenue. However, R² = 0.0056 means")
print("  only 0.56% of revenue variation is explained")
print("  by quantity alone. Product category, pricing,")
print("  and product mix are stronger drivers.")
print("=" * 45)
```

```python
# @title 5. Coefficient of Variation (by Product Category)

cv_df = df[df['Product_Category'] != 'Cook Top']


cv_table = cv_df.groupby('Product_Category')['Total_Sales_Value_(INR)'].agg(['mean', 'std'])
cv_table['CV_%'] = (cv_table['std'] / cv_table['mean']) * 100

cv_table = cv_table.rename(columns={
    'mean': 'Mean Revenue (INR)',
    'std' : 'Standard Deviation'
})

cv_table = cv_table.sort_values(by='CV_%', ascending=False).round(2)

print("Coefficient of Variation — Revenue Variability by Product Category")
print()
display(cv_table)
```

```python
# @title 6. Independent Samples T-Test (FY 2019-20 vs FY 2020-21)
# H0: No significant difference in average sales revenue between FY 2019-20 and FY 2020-21.
# H1: There is a significant difference in average sales revenue between the two fiscal years.

from scipy.stats import ttest_ind

print("Unique Fiscal Years:", df['Fiscal_Year'].unique())
print("Record Count per Year:")
print(df['Fiscal_Year'].value_counts())
print()

fy_2019 = df[df['Fiscal_Year'] == '2019-2020']['Total_Sales_Value_(INR)'].dropna()
fy_2020 = df[df['Fiscal_Year'] == '2020-2021']['Total_Sales_Value_(INR)'].dropna()

t_stat, p_value = ttest_ind(fy_2019, fy_2020, equal_var=False)

print("=" * 45)
print("       INDEPENDENT T-TEST RESULTS")
print("=" * 45)
print(f"  t-statistic : {round(t_stat, 4)}")
print(f"  P-value     : {round(p_value, 6)}")
print("-" * 45)

if p_value < 0.05:
    print("  Result        : Reject H0")
    print("  Interpretation: Significant difference in")
    print("  average sales revenue between FY 2019-20")
    print("  and FY 2020-21.")
else:
    print("  Result        : Accept H0")
    print("  Interpretation: No significant difference")
    print("  in average sales revenue between the")
    print("  two fiscal years.")

print("=" * 45)
```
